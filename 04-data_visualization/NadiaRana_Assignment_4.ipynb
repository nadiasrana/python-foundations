### 2. Create a Python function that calculates the square of a given integer 'x' and returns the result as a floating-point number.

```python
def calculate_square(x):
    # Calculate the square of 'x'
    square = x ** 2

    # Convert the result to a floating-point number
    square_as_float = float(square)

    return square_as_float

# Example usage:
input_number = 5
result = calculate_square(input_number)
print("The square of", input_number, "is", result)
```

### 3. Write a Python function that takes a string as input, removes any whitespace, and returns the string in all lowercase characters. 

```python
def process_string(input_string):
    # Remove whitespace from the input string using the 'replace' method
    no_whitespace_string = input_string.replace(" ", "")

    # Convert the string to lowercase using the 'lower' method
    lowercase_string = no_whitespace_string.lower()

    return lowercase_string

# Example usage:
input_text = "Hello, World!"
processed_text = process_string(input_text)
print("Processed string:", processed_text)
```

### 4. Create a Python function that checks whether a given number 'x' is even and returns boolean type 'True' if it is, or 'False' if it's odd.

```python
def is_even(x):
    # return True if the function input is even
    # return False if the function input is not even
    return x % 2 == 0

# Example usage:
input_number = 10
result = is_even(input_number)

if result:
    print(input_number, "is even.")
else:
    print(input_number, "is odd.")
```
