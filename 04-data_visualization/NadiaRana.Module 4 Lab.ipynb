## Histogram

```python
import pandas as pd
import matplotlib.pyplot as plt

# Generate a sample 'Student' DataFrame with 'Age'
np.random.seed(0)
ages = np.random.randint(18, 30, size=100)  # Generate 100 random ages between 18 and 30
student_ids = range(1, 101)  # Generate student IDs from 1 to 100

# Create the DataFrame
students_df = pd.DataFrame({
    'StudentID': student_ids,
    'Age': ages
})

# Plot the distribution of age
plt.figure(dpi=150)
students_df['Age'].hist(bins=12, edgecolor='black')

plt.title('Distribution of Student Ages')
plt.xlabel('Age')
plt.ylabel('Number of Students')
```

## Density Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Now we plot the density plot for the 'Age' data from the students_df
plt.figure(dpi=150)
sns.kdeplot(students_df['Age'], bw_adjust=0.5, fill=True)

plt.title('Density Plot of Student Ages')
plt.xlabel('Age')
plt.ylabel('Density')
```

## Box Plot

```python

# Generating a boxplot for the 'Age' data from the students_df
plt.figure(dpi=150)
boxplot = sns.boxplot(x=students_df['Age'], color='lightblue', fliersize=5, linewidth=1.5, width = 0.1)

# Calculate Q1, Q3, and median
Q1 = students_df['Age'].quantile(0.25)
Q3 = students_df['Age'].quantile(0.75)
median = students_df['Age'].median()

# Identify the outliers
iqr = Q3 - Q1
outliers = students_df['Age'][((students_df['Age'] < (Q1 - 1.5 * iqr)) |(students_df['Age'] > (Q3 + 1.5 * iqr)))]

# Annotate the Q1, Q3, and median
plt.annotate(f'Q1', xy=(Q1, 0), xytext=(Q1, -0.2),
             arrowprops=dict(facecolor='blue', shrink=0.05), ha='center')
plt.annotate(f'Q3', xy=(Q3, 0), xytext=(Q3, -0.2),
             arrowprops=dict(facecolor='blue', shrink=0.05), ha='center')
plt.annotate(f'Median', xy=(median, 0), xytext=(median, -0.4),
             arrowprops=dict(facecolor='green', shrink=0.05), ha='center')

# Annotate the outliers
for outlier in outliers:
    plt.annotate(f'Outlier', xy=(outlier, 0), xytext=(outlier, 0.3),
                 arrowprops=dict(facecolor='red', shrink=0.05), ha='center')

plt.title('Box Plot of Student Ages')
plt.xlabel('Age')
plt.ylabel('')
```

## Scatter Plot

```python
# Generate a sample dataset with 30 data points for height and weight
np.random.seed(0)  # For reproducibility
heights = np.random.normal(170, 10, 30)  # Average height in cm with some standard deviation
weights = np.random.normal(65, 15, 30)  # Average weight in kg with some standard deviation

# Create a DataFrame from the generated data
body_measurements_df = pd.DataFrame({
    'Height': heights,
    'Weight': weights
})

# Plotting the scatter plot for the generated dataset
plt.figure(dpi=150)
plt.scatter(body_measurements_df['Height'], body_measurements_df['Weight'], color='blue', alpha=0.6)

plt.title('Scatter Plot of Height vs. Weight')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
```

## Pearson correlation coefficient

```python
# Calculate the Pearson correlation coefficient for the height and weight data
correlation_coefficient = body_measurements_df['Height'].corr(body_measurements_df['Weight'])
correlation_coefficient
```

Code Block #1 -- Scatter Plot

```python
# Import required libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load the tips dataset
tips = sns.load_dataset("tips")

# Print first 5 rows to verify dataset
print("First 5 rows of the tips dataset:")
print(tips.head())

# Scatter plot of total_bill vs tip
sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.title("Scatter Plot of Total Bill vs. Tip")
plt.show()
```

### Scatter Plot  Rationale & Insights

This scatter plot visualizes the relationship between total bill amount and tip.The plot shows a positive relationship, meaning that as the total bill increases, the tip amount generally increases as well. This suggests that customers tend to tip proportionally based on how much they spend.

Code Block #2 -- Bar Plot 

```python
print("Average tip by day:")
print(tips.groupby("day", observed=False)["tip"].mean())

sns.barplot(x="day", y="tip", data=tips)

plt.title("Average Tip Amount by Day")

plt.show()

sns.boxplot(x="day", y="total_bill", hue="time", data=tips)

plt.title("Distribution of Total Bill by Day and Time")

plt.show()
```

### Bar and Box Plots  Rationale & Insights

The bar plot shows the average tip amount for each day of the week.
This helps identify which days tend to have higher tipping behavior. The box plot displays the distribution of total bill amounts by day and time (lunch vs dinner). Dinner bills generally show higher variability and higher totals compared to lunch, indicating higher spending during dinner service.

Code Block #3 -- Histogram 

```python
# Create correlation matrix FIRST 
correlation_matrix = tips.corr(numeric_only=True)

print("Summary statistics for total_bill:")
print(tips["total_bill"].describe())

print("Correlation matrix:")
print(correlation_matrix)

sns.histplot(tips["total_bill"], bins=20, kde=True)
plt.title("Histogram of Total Bill")
plt.show()

sns.pairplot(tips, hue="sex")
plt.suptitle("Pair Plot of Numeric Variables (Hue by Gender)")
plt.show()

sns.heatmap(correlation_matrix, annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

### Histogram, Pair Plot, and Heatmap  Rationale & Insights

The histogram shows the distribution of total bill values and indicates that most bills fall within a moderate range, with fewer very high totals. The pair plot allows comparison between numeric variables and highlights relationships such as total bill and tip, separated by gender. The correlation heatmap summarizes relationships between numeric variables. Total bill and tip show a strong positive correlation, confirming earlier observations from the scatter plot.

## Additional Practice Examples

The following visualizations were created as additional practice to further
explore seaborn and matplotlib techniques covered in the course materials.
These examples help reinforce understanding of categorical comparisons,
distributions, and multivariate relationships.

```python
# Violin plot of total bill by day
print("Creating violin plot for total bill by day")

sns.violinplot(x="day", y="total_bill", data=tips)

plt.title("Violin Plot of Total Bill by Day")
plt.show()
```

### Violin Plot  Insights

The violin plot combines features of a box plot and a density plot.
It shows the distribution of total bill amounts for each day, allowing
comparison of spending patterns across days. Some days show wider
distributions, indicating more variability in customer spending.

```python
# Count plot for number of customers by day
print("Counting number of records per day")

sns.countplot(x="day", data=tips)

plt.title("Number of Customers by Day")
plt.show()
```

### Count Plot  Insights

The count plot displays how many observations occur for each day of the week. This helps identify which days have more recorded customer visits and provides context for interpreting other plots such as average tips or total bills.

```python
# Scatter plot with hue for smoker status
print("Scatter plot of total bill vs tip by smoker status")

sns.scatterplot(
    x="total_bill",
    y="tip",
    hue="smoker",
    data=tips
)

plt.title("Total Bill vs Tip by Smoker Status")
plt.show()
```

### Scatter Plot with Hue  Insights

This scatter plot adds smoker status as a third variable using color. It allows comparison of tipping behavior between smokers and non-smokers while still showing the positive relationship between total bill and tip.

```python
# Average total bill by day and time
print("Average total bill grouped by day and time")

sns.barplot(
    x="day",
    y="total_bill",
    hue="time",
    data=tips
)

plt.title("Average Total Bill by Day and Time")
plt.show()
```

### Grouped Bar Plot  Insights

This grouped bar plot compares the average total bill by day, separated
by lunch and dinner. Dinner consistently shows higher average spending,
highlighting how time of day impacts customer behavior.
