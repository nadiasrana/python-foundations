### Code Block #1 — if-elif-else structure (Age categorization)

```python
# Code Block #1: if-elif-else
# Purpose: Categorize a person's age into a label based on sequential conditions.
# Key idea: Python checks conditions from top to bottom; the first True branch runs.

age = 17  # Try changing this value (e.g., 10, 13, 25) and re-run to compare results.

if age < 13:
    category = "child"
elif age < 18:
    category = "teenager"
else:
    category = "adult"

print("Age:", age, "-> Category:", category)
```

**What I observed after changing `age`:**
- If `age` is less than 13 → `child`
- If `age` is 13–17 → `teenager`
- If `age` is 18 or more → `adult`

This helped me see how order matters (the `elif` only runs if the first `if` is false).


### Code Block #2 — for loop (Sum of first n numbers)

```python
# Code Block #2: for loop
# Purpose: Use a for-loop to compute the sum of the first 'n' positive integers.
# Key idea: 'sum_total' updates cumulatively each iteration (running total).

n = 5  # Try changing this value (e.g., 1, 3, 10) and re-run to compare results.

sum_total = 0  # Using a different variable name than 'sum' to avoid shadowing Python's built-in sum().

for i in range(1, n + 1):
    sum_total += i  # Add the current i to the running total

print("n:", n, "-> Sum 1..n =", sum_total)
```

**What I observed after changing `n`:**
- Increasing `n` increases the output because we are adding more numbers.
- The loop iterates exactly `n` times because `range(1, n+1)` includes 1 and excludes `n+1`.


### Code Block #3 — while loop (Print even numbers up to a limit)

```python
# Code Block #3: while loop
# Purpose: Print even numbers from 0 up to a chosen 'limit' using a condition-driven loop.
# Key ideas:
#   1) 'while True' creates an infinite loop
#   2) 'break' stops the loop when a condition is met
#   3) '%' (mod) checks even/odd (even numbers have remainder 0 when divided by 2)

limit = 10  # Try changing this value (e.g., 5, 12, 0) and re-run to compare results.

number = 0

while True:
    # Stop condition: if number exceeds the limit, exit the loop
    if number > limit:
        break

    # Print only even numbers
    if number % 2 == 0:
        print(number)

    # Move to the next number to avoid an infinite loop
    number += 1
```

**What I observed after changing `limit`:**
- The output lists **even numbers** from 0 up to `limit`.
- The `break` statement is what prevents the loop from running forever.

### Extra Example 1 — if-elif-else (Grade bucket)

```python
# Purpose: Bucket a score into a letter grade using if-elif-else.
# This mirrors the same decision logic as the age categorization.

score = 86  # Try: 59, 60, 72, 95

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"

print("Score:", score, "-> Grade:", grade)
```

### Extra Example 2 — for loop (Average of a list)

```python
# Purpose: Use a for-loop to compute an average from a list of numbers.

values = [10, 15, 25, 50]
total = 0

for v in values:
    total += v

average = total / len(values)
print("Values:", values, "-> Average:", average)
```

### Extra Example 3 — while loop (Countdown)

```python
# Purpose: Use a while loop with a clear stopping condition to count down to 0.

count = 5  # Try changing this value

while count >= 0:
    print(count)
    count -= 1

print("Done!")
```
